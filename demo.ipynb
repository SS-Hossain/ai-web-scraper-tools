```python
import pandas as pd
import requests
import json

# Function to load and display scraped data
def load_csv(file_path):
    try:
        df = pd.read_csv(file_path)
        display(df.head())  # Show the first 5 rows
    except Exception as e:
        print(f"Error loading CSV: {e}")

# Function to summarize data using GPT API (Example placeholder)
def summarize_data(text):
    api_key = "your-serpapi-key-here"  # Replace with your SerpAPI key
    # url = "https://api.example.com/summarize"  # Placeholder API endpoint - This is likely incorrect
    url = "https://serpapi.com/search"  # Assuming you intend to use SerpAPI for summarization as well

    payload = {
        "text": text,
        "key": api_key,
        "engine": "google"  # Add the search engine parameter
    }

    response = requests.post(url, json=payload)
    # Check if the response was successful before trying to parse JSON
    if response.status_code == 200:
        data = response.json()
        # Assuming SerpAPI response structure
        return data.get("answer_box", {}).get("snippet") or \
            data.get("organic_results", [{}])[0].get("snippet", "No summary available")
    else:
        return f"Error in summarization API request (Status Code: {response.status_code})"

# Load and display Google Search results
print("\nLoading Google Search Results...")
load_csv("/google_search_results.csv")

# Load and display YouTube Video results
print("\nLoading YouTube Video Results...")
load_csv("/youtube_video_results.csv")

# Example text for summarization
demo_text = "This is a sample text that we want to summarize using AI-powered summarization."
print("\nSummary:", summarize_data(demo_text))
```
